# Extracting Genres feature using TMDB API

```python
import numpy as np
import pandas as pd
from tmdbv3api import TMDb,Movie
import json
import requests
```

```python
tmdb=TMDb()
tmdb.api_key='ff5d4bbd48363bb3f65e4c34e204d94b'
tmdb_movie=Movie()
def get_genre(x):
    genres=[]
    result=tmdb_movie.search(x)
    movie_id=result[0].id
    response=requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json=response.json()
    if data_json['genres']:
        genre_str=" " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        np.NaN
```

## Extracting Genres of 2018 movies

```python
df1=pd.read_csv('2018.csv')
df1.head()
```

```python
df1.drop(['Unnamed: 0','Unnamed: 6','Opening','Opening.1','Ref.'],axis=1,inplace=True)
df1.head()
```

```python
df1['genres'] = df1['Title'].map(lambda x: get_genre(str(x)))
df1.head()
```

```python
df1.to_csv('2018_with_genres.csv')
```
